1] What is the difference between a function and a method in Python?
- Function: A block of code that performs a task and is defined using def. It can exist independently.ex-
      def greet():  
        print("Hello")

Method: A function is associated with an object (usually a class). Called on objects. ex-

      class Person:  
        def greet(self):  
          print("Hello")

2] Explain the concept of function arguments and parameters in Python.
- Parameter: A variable in the function definition.They are the placeholders.ex-

      def greet(name):  # 'name' is a parameter
        print("Hello", name)

Argument: It is the actual value passed to a function when calling it.
ex-

    greet("abish")  # "abish" is an argument

3] What are the different ways to define and call a function in Python?
- Ways to define a function:
  1. Standard function (def)
  2. Lambda (anonymous) function
  3. Generator Function (yield)
  4. Nested Function
  5. Recursive Function
  6. Function Returning Function
  7. Method (Class Function)
  8. Callable Class (__call__)
  9. Decorator Function

Ways to call a function:
  1. Basic Call: func(arg1, arg2)


  2. Keyword Arguments: func(name="Alice", age=25)


  3. Default Arguments: func()


  4. Variable-Length Arguments

    *args: func(1, 2, 3) (positional → tuple)

    **kwargs: func(name="Alice", age=25) (keyword → dict)


  5. Lambda Call: (lambda x: x+1)(5)


  6. Recursive Call: factorial(n-1) (function calls itself)


  7. Higher-Order Function Call: map(func, iterable)


  8. Method Call (OOP): obj.method()


  9. Callable Class (__call__): obj() (if class defines __call__)


  10. Decorator Call: @decorator modifies func() behavior


  11. List Comprehension: [func(x) for x in iterable]


  12. Argument Unpacking

     func(*args) (unpacks list/tuple)    

     func(**kwargs) (unpacks dict)


  13. Partial Function (functools.partial): partial(func, arg1)(arg2)


  14. Threaded/Async Call: threading.Thread(target=func).start()


  15. Dynamic Import + Call: getattr(module, "func_name")()

6] Explain the concept of generators in Python and how they are defined.
- Generator: A special type of iterator that yields values lazily (one at a time) using yield.
They are defined using a function with yield instead of return.
They are efficient for large data and they doesn't store all values in memory.

7] What are the advantages of using generators over regular functions?
- Advantages of generators:

    - Memory efficient - yield items one at a time, not all at once.
    
    - Faster startup - execution starts immediately, not after full list creation.

    - Infinite sequences - can model endless data streams.

    - Simpler code - avoids manual iterator creation.

9] Explain the purpose and usage of the `map()` function in Python.
- The map() function applies a given function to each item in an iterable (e.g., list, tuple) and returns an iterator of results.

Key Features:

  - Lazy Evaluation: Returns an iterator (use list() to get actual values).

  - Efficiency: Processes items one-by-one without creating intermediate lists.

  - Flexibility: Works with any iterable and any function (built-in, lambda, or custom).

Common use cases are to transform all items, in type conversion, in multiple iterables.

10] What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map(function, iterable):

      Purpose: Transforms each item in an iterable by applying a function.

      Returns: An iterator of transformed items (same length as input).

 filter(function, iterable):

      Purpose: Selects items from an iterable where the function returns True.

      Returns: An iterator of filtered items (length ≤ input).

 reduce(function, iterable):

      Purpose: Aggregates all items into a single value by cumulative computation.

      Returns: A single combined result (not an iterator).

```python
#1]  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
#4] Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Test for numbers 1 to 200
for num in range(1, 201):
    if prime(num):
        print(f"{num} is prime")
    else:
        print(f"{num} is not prime")
```

```python
#5]  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
#6] Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power)
```

```python
#7]  Implement a generator function that reads a file line by line and yields each line as a string
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
#8] Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 3), (4, 1), (2, 5), (3, 2)]

# Sort by second element
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
#9] Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# Celsius to Fahrenheit conversion: F = (C * 9/5) + 32

celsius_temps = [0, 20, 37, 100]

fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print(fahrenheit_temps)
```

```python
#10]  Create a Python program that uses `filter()` to remove all the vowels from a given string.
input_str = input("Enter a string: ")

# Define vowels
vowels = "aeiouAEIOU"

# Use filter to exclude vowels
result = ''.join(filter(lambda ch: ch not in vowels, input_str))

print("String without vowels:", result)
```
